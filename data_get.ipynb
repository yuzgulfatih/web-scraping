```python
import json
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# JSON dosyalarının bulunduğu klasör
klasor_yolu = "Script Jsons"

# WebDriver'ı başlat
chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Klasördeki tüm JSON dosyalarını oku
    for dosya_adi in os.listdir(klasor_yolu):
        if dosya_adi.endswith(".json"):
            dosya_yolu = os.path.join(klasor_yolu, dosya_adi)

            with open(dosya_yolu, "r+", encoding='utf-8') as dosya:
                veri = json.load(dosya)
                imdb_url = veri.get("imdb_url")

                if imdb_url:
                    if not imdb_url.endswith("/"):
                        imdb_url += "/"

                    # Sayfaya git
                    driver.get(imdb_url)

                    # Sayfanın tamamen yüklenmesini bekle
                    time.sleep(5)  # İlk yüklenme için bekle

                    # Sayfanın en aşağısına in
                    scroll_pause_time = 2  # İhtiyaca göre süreyi değiştirebilirsin
                    last_height = driver.execute_script("return document.body.scrollHeight")

                    while True:
                        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                        time.sleep(scroll_pause_time)
                        new_height = driver.execute_script("return document.body.scrollHeight")
                        if new_height == last_height:
                            break
                        last_height = new_height

                    # Sayfanın en aşağısına inildiği için yukarıya yavaş yavaş çık
                    while True:
                        driver.execute_script("window.scrollBy(0, -window.innerHeight);")
                        time.sleep(scroll_pause_time)
                        scroll_position = driver.execute_script("return window.pageYOffset;")
                        if scroll_position == 0:
                            break

                        # HTML kaynağını tekrar al ve kontrol et
                        html_source = driver.page_source
                        soup = BeautifulSoup(html_source, 'html.parser')

                        # Storyline bölümünü bul
                        storyline_section = soup.find('section', {'data-testid': 'Storyline'})

                        if storyline_section:
                            # İlgili div'in içeriğini al
                            storyline_text = storyline_section.find('div', {'class': 'ipc-html-content-inner-div'}).get_text(strip=True)
                            veri['description'] = storyline_text
                            print(f"{dosya_adi} için storyline eklendi:", storyline_text)
                            break

                    # JSON dosyasını güncelle
                    dosya.seek(0)
                    json.dump(veri, dosya, ensure_ascii=False, indent=4)
                    dosya.truncate()

finally:
    # Tarayıcıyı kapat
    driver.quit()
```

```python
import json
import os
import requests
from bs4 import BeautifulSoup

# JSON dosyalarının bulunduğu klasör
klasor_yolu = "Script Jsons"

# Headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

try:
    # Klasördeki tüm JSON dosyalarını oku
    for dosya_adi in os.listdir(klasor_yolu):
        if dosya_adi.endswith(".json"):
            dosya_yolu = os.path.join(klasor_yolu, dosya_adi)

            with open(dosya_yolu, "r+", encoding='utf-8') as dosya:
                veri = json.load(dosya)
                imdb_url = veri.get("imdb_url")

                if imdb_url:
                    if not imdb_url.endswith("/"):
                        imdb_url += "/"

                    try:
                        # Sayfayı getir
                        response = requests.get(imdb_url, headers=headers)
                        response.raise_for_status()  # Hata kontrolü

                        # HTML kaynağını al
                        soup = BeautifulSoup(response.content, 'html.parser')

                        # İlk önce sc-d8941411-1 fTeJrK sınıfını ara
                        title_div = soup.find('div', {'class': 'sc-d8941411-1 fTeJrK'})
                        if title_div:
                            original_title = title_div.get_text(strip=True)
                        else:
                            # Eğer sc-d8941411-1 fTeJrK sınıfı yoksa hero__primary-text sınıfını ara
                            title_span = soup.find('span', {'class': 'hero__primary-text'})
                            if title_span:
                                original_title = title_span.get_text(strip=True)
                            else:
                                original_title = None

                        if original_title:
                            veri['name'] = original_title

                            # JSON dosyasını güncelle
                            dosya.seek(0)
                            json.dump(veri, dosya, ensure_ascii=False, indent=4)
                            dosya.truncate()
                            print(f"{dosya_adi} için original title:", original_title)
                        else:
                            print(f"{dosya_adi} için original title bulunamadı")

                    except requests.exceptions.RequestException as e:
                        print(f"{dosya_adi} için HTTP Hatası: {e}")
                    except Exception as e:
                        print(f"{dosya_adi} için Hata: {e}")

except Exception as e:
    print(f"Genel hata: {e}")
```

```python
import os
import json

# "Movies" klasörünü oluştur
movies_klasoru = "Movies"
if not os.path.exists(movies_klasoru):
    os.makedirs(movies_klasoru)

# JSON dosyalarının bulunduğu klasör
klasor_yolu = "Script Jsons/"

# Klasördeki tüm JSON dosyalarını işle
for dosya_adı in os.listdir(klasor_yolu):
    if dosya_adı.endswith(".json"):
        dosya_yolu = os.path.join(klasor_yolu, dosya_adı)
        try:
            # JSON dosyasını yükle
            with open(dosya_yolu, "r+") as dosya:
                veri = json.load(dosya)
                
                # JSON verisinden gerekli alanları al
                id = veri.get("imdb_id")
                description = veri.get("description", "")
                rate = veri.get("imdb_rating", "")
                name = veri.get("name", "")
                genres = ", ".join(veri.get("genres", []))
                imageUrl = veri.get("imdb_image_url", "")
                
                # Oluşturulacak yeni JSON verisi
                yeni_veri = {
                    "id": id,
                    "name": name,
                    "description": description,
                    "genre": genres,
                    "imageUrl": imageUrl,
                    "rate": rate
                }
                
                # "Movies" klasöründe yeni JSON dosyası oluştur
                yeni_dosya_yolu = os.path.join(movies_klasoru, dosya_adı)
                with open(yeni_dosya_yolu, "w") as yeni_dosya:
                    json.dump(yeni_veri, yeni_dosya, indent=4)
        except json.JSONDecodeError as e:
            print(f"{dosya_yolu} dosyası geçersiz JSON formatına sahip: {str(e)}")
```

```python
import os
import json
import re

# JSON dosyalarının bulunduğu klasör
klasor_yolu = "Script Jsons/"

# Klasördeki tüm JSON dosyalarını işle
for dosya_adı in os.listdir(klasor_yolu):
    if dosya_adı.endswith(".json"):
        dosya_yolu = os.path.join(klasor_yolu, dosya_adı)
        try:
            # JSON dosyasını yükle
            with open(dosya_yolu, "r+", encoding='utf-8') as dosya:
                veri = json.load(dosya)
                
                # "dialogues" alanını kontrol et
                if "dialogues" in veri:
                    # "dialogues" içindeki metinleri güncelle
                    updated_dialogues = {}
                    for character, dialogue in veri["dialogues"].items():
                        # \t ve \n karakterlerini kaldır
                        updated_dialogue = dialogue.replace('\t', '').replace('\n', ' ')
                        # Köşeli parantezler içindeki sayıları kaldır
                        updated_dialogue = re.sub(r'\[\d+\]', '', updated_dialogue)
                        updated_dialogues[character] = updated_dialogue.strip()
                    
                    # Güncellenmiş "dialogues" verisini JSON dosyasına yaz
                    veri["dialogues"] = updated_dialogues
                    dosya.seek(0)
                    json.dump(veri, dosya, ensure_ascii=False, indent=4)
                    dosya.truncate()
        
        except json.JSONDecodeError as e:
            print(f"{dosya_yolu} dosyası geçersiz JSON formatına sahip: {str(e)}")
        except Exception as e:
            print(f"Hata: {str(e)}")
```

```python
import os
import json

# JSON dosyalarının bulunduğu klasör
klasor_yolu = "Script Jsons/"
output_klasoru = "Dialogues"

# Çıktı klasörünü oluştur
os.makedirs(output_klasoru, exist_ok=True)

# Toplanacak veriler için boş bir liste
toplanan_veriler = []

# Klasördeki tüm JSON dosyalarını işle
for dosya_adı in os.listdir(klasor_yolu):
    if dosya_adı.endswith(".json"):
        dosya_yolu = os.path.join(klasor_yolu, dosya_adı)
        try:
            # JSON dosyasını yükle
            with open(dosya_yolu, "r", encoding='utf-8') as dosya:
                veri = json.load(dosya)
                
                # Gerekli alanları kontrol et ve topla
                imdb_id = veri.get("imdb_id")
                dialogues = veri.get("dialogues")
                
                if imdb_id and dialogues:
                    toplanan_veriler.append({
                        "id": imdb_id,
                        "dialogues": dialogues
                    })
        
        except json.JSONDecodeError as e:
            print(f"{dosya_yolu} dosyası geçersiz JSON formatına sahip: {str(e)}")
        except Exception as e:
            print(f"Hata: {str(e)}")

# Toplanan verileri tek bir JSON dosyasında birleştir
output_dosyasi = os.path.join(output_klasoru, "id-script.json")
with open(output_dosyasi, "w", encoding='utf-8') as output_dosya:
    json.dump(toplanan_veriler, output_dosya, ensure_ascii=False, indent=4)

print(f"Tüm veriler '{output_dosyasi}' dosyasına kaydedildi.")
```

```python
import json

def split_and_combine_dialogues(input_file, output_file, max_length):
    # JSON dosyasını okumak
    with open(input_file, 'r') as file:
        data = json.load(file)
    
    # Her bir diyalog içindeki uzun metinleri parçalamak
    for item in data:
        for key, dialogue in item.get("dialogues", {}).items():
            if len(dialogue) > max_length:
                # Metni belirli uzunlukta parçalara bölmek
                parts = [dialogue[i:i + max_length] for i in range(0, len(dialogue), max_length)]
                # Parçaları birleştirerek tek bir metin haline getirmek
                combined_dialogue = '\n'.join(parts)
                item["dialogues"][key] = combined_dialogue

    # Parçalanmış ve birleştirilmiş veriyi tek bir dosyaya yazmak
    with open(output_file, 'w') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
    print(f'{output_file} yazıldı.')

if __name__ == "__main__":
    input_file = 'Dialogues/id-script.json'
    output_file = 'output_combined.json'
    max_length = 1800  # Metin parçalarının maksimum uzunluğu
    split_and_combine_dialogues(input_file, output_file, max_length)
```

```python
import json
import re

def split_text_at_period(text, max_length):
    """ Metni max_length sınırında böl ve en uygun noktaya kadar kes. """
    if len(text) <= max_length:
        return [text]
    
    # max_length karakterden kesilmemiş en uygun nokta '.' bul
    split_positions = [m.start() for m in re.finditer(r'\.', text[:max_length+1])]
    if not split_positions:
        # Eğer nokta bulunmazsa, doğrudan max_length'ye böl
        split_point = max_length
    else:
        split_point = split_positions[-1] + 1

    return [text[:split_point]] + split_text_at_period(text[split_point:].lstrip(), max_length)

def combine_and_split_dialogues(input_file, output_file, max_length):
    # JSON dosyasını okumak
    with open(input_file, 'r') as file:
        data = json.load(file)

    # Her bir diyalog içindeki metinleri birleştirerek ve bölerek tek bir metin haline getirmek
    for item in data:
        dialogues = item.get("dialogues", {})
        combined_dialogue = ", ".join(dialogues.values())
        split_dialogues = split_text_at_period(combined_dialogue, max_length)
        item["dialogues"] = split_dialogues

    # Parçalanmış ve birleştirilmiş veriyi tek bir dosyaya yazmak
    with open(output_file, 'w') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
    print(f'{output_file} yazıldı.')

if __name__ == "__main__":
    input_file = 'Dialogues\id-script.json'
    output_file = 'output_combined_split.json'
    max_length = 1800 # Metin parçalarının maksimum uzunluğu
    combine_and_split_dialogues(input_file, output_file, max_length)
```

```python
import json
import re

def split_into_sentences(text):
    # Metni noktalar, ünlemler ve soru işaretleri ile böl
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', text)
    return sentences

def combine_and_split_dialogues(input_file, output_file):
    # JSON dosyasını okumak
    with open(input_file, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Her bir diyalog içindeki metinleri birleştirerek ve cümlelere ayırarak tek bir metin haline getirmek
    for item in data:
        dialogues = item.get("dialogues", [])
        combined_dialogue = " ".join(dialogues)
        split_dialogues = split_into_sentences(combined_dialogue)
        item["dialogues"] = split_dialogues

    # Parçalanmış ve birleştirilmiş veriyi tek bir dosyaya yazmak
    with open(output_file, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
    print(f'{output_file} yazıldı.')

if __name__ == "__main__":
    input_file = 'Dialogues/id-script.json'
    output_file = 'output_combined_split.json'
    combine_and_split_dialogues(input_file, output_file)
```

```python
import json
import re

def split_into_sentences(text):
    # Metni noktalar, ünlemler ve soru işaretleri ile böl ve trim yap
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', text)
    return [sentence.strip() for sentence in sentences]

# JSON dosyasını oku
with open('Dialogues/output_combined_split.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# JSON verisinin bir liste olup olmadığını kontrol et
if isinstance(data, list):
    new_data_list = []
    for item in data:
        new_data = {"id": item["id"], "dialogues": []}
        for dialogue in item['dialogues']:
            sentences = split_into_sentences(dialogue)
            new_data['dialogues'].extend(sentences)  # Append yerine extend kullanarak düz liste oluştur
        new_data_list.append(new_data)
    # Yeni JSON yapısını dosyaya yaz
    with open('output.json', 'w', encoding='utf-8') as file:
        json.dump(new_data_list, file, ensure_ascii=False, indent=4)
else:
    # Tek bir sözlük varsa, işlemi doğrudan gerçekleştir
    new_data = {"id": data["id"], "dialogues": []}
    for dialogue in data['dialogues']:
        sentences = split_into_sentences(dialogue)
        new_data['dialogues'].extend(sentences)  # Append yerine extend kullanarak düz liste oluştur
    # Yeni JSON yapısını dosyaya yaz
    with open('output.json', 'w', encoding='utf-8') as file:
        json.dump(new_data, file, ensure_ascii=False, indent=4)

print("Yeni JSON dosyası oluşturuldu: output.json")
```

Json to csv

```python
import os
import pandas as pd
import json

# JSON dosyalarının bulunduğu klasör
klasor_yolu = 'Script Jsons/'

# Tüm JSON dosyalarının içeriğini birleştirmek için boş bir liste oluşturalım
json_verileri = []

# Dosya adlarını ve içeriklerini birleştir
for dosya_adı in os.listdir(klasor_yolu):
    if dosya_adı.endswith('.json'):
        dosya_yolu = os.path.join(klasor_yolu, dosya_adı)
        
        # JSON dosyasını aç ve içeriğini birleştir
        with open(dosya_yolu, 'r', encoding='utf-8') as dosya:
            json_veri = json.load(dosya)
            json_veri['filename'] = dosya_adı  # Dosya adını sütun olarak ekle
            json_verileri.append(json_veri)

# Tüm JSON verilerini içeren bir DataFrame oluşturalım
df = pd.DataFrame(json_verileri)

# CSV dosyasına yazma
csv_dosya_yolu = 'data/veri.csv'
df.to_csv(csv_dosya_yolu)
```
